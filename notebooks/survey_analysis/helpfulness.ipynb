```python
import numpy as np

#Helpfulness scores of scenarios 5-8 from survey B
s1 = np.array([2, 5, 9, 6, 8])
s2 = np.array([3, 5, 10, 8, 4])
s3 = np.array([3, 5, 4, 8, 10])
s4 = np.array([2, 5, 12, 9, 2])

#Helpfulness scores of scenarios 5-8 from survey A
s5 = np.array([4, 2, 3, 10, 15])
s6 = np.array([0, 3, 1, 16, 14])
s7 = np.array([0, 3, 6, 12, 13])
s8 = np.array([1, 7, 7, 11, 8])
```

```python
m = np.array([s1, s2, s3, s4, s5, s6, s7, s8])
m
```

```python
m_alt = np.array([s1/30, s2/30, s3/30, s4/30, s5/34, s6/34, s7/34, s8/34])
m_alt
```

```python
m_alt.sum(axis=1) #each scenario has 100% responses, hence corrected for 34 vs 30 responses.
```

```python
scores_total_corrected = np.sum(m_alt, axis=0)/np.sum(np.sum(m_alt, axis=0))*100
```

```python
import matplotlib.pyplot as plt
 
# create data
likert_labels = ['1', '2', '3', '4', '5']

# plot bars in stack manner
plt.bar(likert_labels, scores_total_corrected, color='b')
plt.xlabel('Likert scale')
plt.ylabel('Percentage of responses')
plt.title('Helpfulness all scenarios')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example matrix (replace this with your actual 8x5 matrix)
matrix = m_alt*100

# Plotting the bar chart
x = np.arange(1, 6)  # x-axis labels from 1 to 5
width = 0.1  # width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Plot each scenario's bars
for i in range(matrix.shape[0]):
    ax.bar(x + i*width, matrix[i], width, label=f'Scenario {i+1}')

# Adding labels and title
ax.set_xlabel('Likert values')
ax.set_ylabel('Votes (as % by scenario)')
ax.set_title('VA helpfulness contributions of Likert values in different scenarios')
ax.set_xticks(x + width * (matrix.shape[0] - 1) / 2)
ax.set_xticklabels([f'{i}' for i in range(1, 6)])
ax.legend()

plt.show()
```
