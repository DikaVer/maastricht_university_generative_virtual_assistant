```python
import numpy as np

#Response time scores of VA in scenarios 5-8 from survey B
response_time_1 = np.array([0, 3, 6, 17, 4])

#Response time scores of VA in scenarios 5-8 from survey A
response_time_2 = np.array([0, 2, 6, 13, 13])
```

```python
m_alt = np.array([response_time_1/30, response_time_2/34])
m_alt
```

```python
m_alt.sum(axis=1) #each scenario has 100% responses, hence corrected for 34 vs 30 responses.
```

```python
scores_total_corrected = np.sum(m_alt, axis=0)/np.sum(np.sum(m_alt, axis=0))*100
```

```python
import matplotlib.pyplot as plt
 
# create data
likert_labels = ['1', '2', '3', '4', '5']

# plot bars in stack manner
bars = plt.bar(likert_labels, scores_total_corrected, color='b')
plt.xlabel('Likert scale')
plt.ylabel('Percentage of responses')
plt.title('Valuableness of VA system by students')
plt.show()
```

```python
bars = plt.bar(likert_labels, scores_total_corrected, color='b')


# Customizing grid lines
plt.grid(color='gray', linestyle=':', linewidth=0.5, zorder=0)
plt.gca().set_axisbelow(True)

# Creating labels for bars
for bar in bars:
    height = bar.get_height()
    height = round(height, 2)
    plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom')

plt.xlabel('Likert scale')
plt.ylabel('Percentage of responses')
plt.title('Valuableness of VA system by students')
plt.show()
```
