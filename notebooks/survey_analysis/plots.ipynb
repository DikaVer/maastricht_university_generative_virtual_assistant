```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the Bachelor's Program question
program_labels = ['CS', 'DS & AI']
program_total = [13 + 15, 21 + 15]  # Summing survey A and survey B for each category

# Data for the Currently Enrolled Year question
year_labels = ['First Year', 'Second Year', 'Third Year']
year_total = [30 + 24, 2 + 4, 2 + 1]  # Summing survey A and survey B for each category

# Data for the Read the R&R Before? question
rr_labels = ['Yes', 'No']
rr_total = [20 + 15, 14 + 15]  # Summing survey A and survey B for each category

# Combine all data
labels = program_labels + year_labels + rr_labels
survey_total = program_total + year_total + rr_total

# Number of categories
num_categories = len(program_labels) + len(year_labels) + len(rr_labels)

# X positions for categories
x = np.arange(num_categories)

# Width of the bars
width = 0.5

# Define colors for each group of bars
colors = ['skyblue', 'dodgerblue', 'lightgreen', 'mediumseagreen', 'darkgreen', 'lightcoral', 'indianred']

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Plotting the data with specified colors and no edgecolor
bars_total = ax.bar(x, survey_total, width, color=colors, edgecolor='none')

# Adding labels, title, and legend
ax.set_xlabel('Categories', fontsize=12)
ax.set_ylabel('Number of Students', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=10)

# Adding grid lines for better readability with lighter lines
ax.yaxis.grid(True, linestyle='--', which='both', color='gray', alpha=0.3)

# Function to add value labels on the bars
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adding value labels to the bars
autolabel(bars_total)

# Adding horizontal lines to separate sections
ax.axvline(len(program_labels) - 0.5, color='gray', linestyle='--', linewidth=1, alpha=0.5)
ax.axvline(len(program_labels) + len(year_labels) - 0.5, color='gray', linestyle='--', linewidth=1, alpha=0.5)

# Add group labels above the bars with adjusted vertical offset
ax.text((len(program_labels) - 1) / 2, max(survey_total) + 5, 'Bachelor\'s Program', ha='center', fontsize=12, fontweight='bold')
ax.text(len(program_labels) + (len(year_labels) - 1) / 2, max(survey_total) + 5, 'Currently Enrolled Year', ha='center', fontsize=12, fontweight='bold')
ax.text(len(program_labels) + len(year_labels) + (len(rr_labels) - 1) / 2, max(survey_total) + 5, 'Read the R&R Before?', ha='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Helpfulness scores of scenarios 5-8 from survey B e A combinati
s1 = np.array([2, 5, 9, 6, 8])
s2 = np.array([3, 5, 10, 8, 4])
s3 = np.array([3, 5, 4, 8, 10])
s4 = np.array([2, 5, 12, 9, 2])
s5 = np.array([4, 2, 3, 10, 15])
s6 = np.array([0, 3, 1, 16, 14])
s7 = np.array([0, 3, 6, 12, 13])
s8 = np.array([1, 7, 7, 11, 8])
m_helpfulness = s1 + s2 + s3 + s4 + s5 + s6 + s7 + s8

# Response time scores combinati
response_time_1 = np.array([1, 6, 12, 6, 5])
response_time_2 = np.array([1, 5, 12, 11, 5])
m_response_time = response_time_1 + response_time_2

# Valuableness scores combinati
valuable_1 = np.array([0, 3, 6, 17, 4])
valuable_2 = np.array([0, 2, 6, 13, 13])
m_valuable = valuable_1 + valuable_2

# Preparazione del dataframe
data = {
    'Helpfulness': m_helpfulness,
    'Response Time': m_response_time,
    'Valuableness': m_valuable
}

df = pd.DataFrame(data, index=["1", "2", "3", "4", "5"]).reset_index().melt(id_vars=['index'], var_name='Category', value_name='Count')
df.columns = ['Likert', 'Category', 'Count']

# Calcolo delle percentuali
df['Percentage'] = df.groupby('Category')['Count'].transform(lambda x: x / x.sum() * 100)

# Creazione del grafico a barre impilate divergente
categories = df['Category'].unique()
likerts = ["1", "2", "3", "4", "5"]
colors = ["#d6604d", "#f4a582", "#999999", "#92c5de", "#4393c3"]

fig, ax = plt.subplots(figsize=(10, 6))

for i, category in enumerate(categories):
    sub_df = df[df['Category'] == category]
    prev = np.zeros(len(likerts))
    for j, likert in enumerate(likerts):
        values = sub_df[sub_df['Likert'] == likert]['Percentage'].values
        ax.barh(i, values, left=prev, color=colors[j], edgecolor='white', label=likert if i == 0 else "")
        prev += values

# Configurazione dell'asse e delle etichette
ax.set_yticks(np.arange(len(categories)))
ax.set_yticklabels(categories)
ax.set_xlabel('Percentage')
ax.set_title('VA System Feedback')
ax.legend(title="Likert scale", bbox_to_anchor=(1.05, 1), loc='upper left')

# Limiti dell'asse x
ax.set_xlim(0, 100)

plt.show()
```
